#### 1.4. Datatypes analysis

```python
df.info()
```

- Seems like most columns have already been inferred a suitable datatype (by the _pd.read\_csv_ function).

- However, there are some columns (`children`, `agent`, `company`) that have been inferred as float datatype while they should have been integer type (their values are integers in the original dataset). The reason for this is that in default, Pandas represent a missing value as a special floating point number. Therefore, columns containing missing values will be inferred as float type instead. We'll handle this case in the next section.

#### 1.5. Missing value analysis

Columns with missing values

```python
total_missing = df.isna().sum()
total_missing = total_missing[total_missing != 0] # filter columns with no missing values

pd.DataFrame({"Total missing": total_missing, 
              "Missing percentage (%)": total_missing / df.shape[0] * 100})
```

- Looks like the `company` attribute is having a big problem since it's missing in approximately __94.3%__ total of records. Therefore, we can simply just remove this column from our analysis since it won't provide any big helps, and filling or inferring such a large amount of missing values is almost impossible.

- For the `agent` attribute, with nearly of __13.7%__ missing percentage, we can fill in these missing values instead of just dropping them since __13.7%__ can be considered as acceptable in this case.

- For `country` and `children` attributes, their missing percentage is negligible (less than __1%__), so we'll just fill in these missing values with some arbitraries to handle it.

#### 1.6. Handle missing value

- Drop `company` attribute

```python
df = df.drop(columns = ['company'])
```

- For `agent` attribute, to retain its original datatype, we will just fill in its missing values by a special integer (or an agent ID) that doesn't exist in any non-missing rows (which is __0__ in this case).

- Since `children` attribute represents the number of children, __0__ will also be a suitable representation for missing rows.

- For `country`, we'll simply just replace its missing values as __Unknown__.

```python
print("Number of agents having ID 0:", df[df['agent'] == 0].shape[0])
df = df.fillna({
    'agent': 0,
    'children' : 0,
    'country': 'Unknown',
})
```

We also need to convert the datatype of some columns (`agent` and `children`) back to integer to handle them easier later (as mentioned above).

```python
df[['children', 'agent']] = df[['children', 'agent']].astype(int)

df.info()
```

Let's check for missing values again

```python
any(df.isna().sum())
```

#### 1.7. Handle duplicated values

```python
df.duplicated().sum()
```

As we can see, the data have `31987` duplicate entries in the dataset, so let's drop them from our data.

```python
df = df.drop_duplicates()
```

Also, during the process of observing the data, we've discovered several lines of data that have `adult = 0, children = 0, and babies = 0`. Such records can be discarded since a hotel reservation shouldn't be booked with no guest.

```python
df = df[~((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0))]
```

#### 2.2. Visualization of Categorical Data Distribution

First, we will draw pie charts to visualize the distribution of categorical variables. 

_(Except for some variables, such as `country`, `month`... that have too many unique values that shouldn't be put into a single pie chart, we'll analyze them separately later using another plot type)_

```python
# Exclude specified columns
sub_categorical_df = categorical_df.drop(columns=['reservation_status_date', 'arrival_date_month', 'country'])

# Determine the layout for the subplots
n = len(sub_categorical_df.columns)
ncols = 3
nrows = n // ncols if n % ncols == 0 else n // ncols + 1

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(22, 6*nrows))
fig.suptitle("Categorical variables distribution", y=1.02,fontsize=30)
i = 0
for ax in axes.flatten():
    if i >= n:
        ax.set_visible(False)
        continue
    column = sub_categorical_df.columns[i]
    pie_data = sub_categorical_df[column].value_counts()
    pie_data.plot(kind='pie', autopct='%1.1f%%', ax=ax)
    ax.set_title(f'{column}', fontsize=18, fontweight="bold")
    ax.set_ylabel('')
    ax.legend(title=column, bbox_to_anchor=(1,1))  # Add a legend for each pie chart
    
    i += 1

plt.tight_layout()
plt.show()
```

**Comment:** 
- We can see that most variables exhibit a skewed distribution towards a specific value that almost dominates its counterpart. Specifically, for `deposit_type`, most reservations aren't booked with deposit, most guests choose BB `meal` type, most are transient (`customer_type`), etc.

- Also, from `is_repeated_guest` variable, we can see that only about __4%__ of guests come back to same hotel and it's also noteworthy to know that the cancellation rate (`is_canceled`) is pretty high in this case, which is about __27.5%__ reservations are canceled.

#### 3.2. Using scatter plot to analyze relationships of variables

##### Booking Lead Time vs Average Daily Rate
Let's see how longer lead time can affect the revenue of the hotel by showing the relationship between them.

```python
d = df.groupby(['hotel','lead_time']).agg({'adr': 'mean'}).rename(columns={'adr':'mean_adr'}).reset_index()
```

```python
plt.figure(figsize=(9,6))
sns.scatterplot(x='lead_time', y='mean_adr', hue='hotel', data=d)
plt.title("Booking Lead Time vs Average Daily Rate", fontweight="bold")
sns.regplot(x='lead_time', y='mean_adr', data=d[d['hotel']=='City Hotel'])
sns.regplot(x='lead_time', y='mean_adr', data=d[d['hotel']=='Resort Hotel'])
plt.xlabel("Days")
plt.ylabel("Average daily rate")
plt.show()
```

**Comment:** 
- There appears to be a negative correlation between booking lead time and average daily rate. As the `booking lead time` increases, the `average daily rate` tends to decrease. This trend is indicated by the downward slope of the regression lines for both hotel types (City Hotel and Resort Hotel).

- However, the data points for the Resort Hotel seem to be more spread out, indicating a wider range of average daily rates compared to the City Hotel. This variability is expected in real-world hotel booking data due to factors such as seasonality, location, and demand fluctuations.

##### Effect of `Lead time` on `Cancellations`

```python
d = df.groupby('lead_time').agg({'is_canceled':'sum', 'hotel':'count'}).reset_index().rename(columns={'is_canceled':'cancelled_bookings', 'hotel':'total_bookings'})
d['cancellation_percentage'] = (d['cancelled_bookings']/d['total_bookings'])*100
d.head()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='lead_time', y='cancellation_percentage', data=d)
plt.ylim((0,100))
plt.xlabel("Lead Time (days)")
plt.ylabel("% Cancellation")
plt.title("Effect of Lead time on Cancellations", fontweight="bold")
plt.show()
```

**Comment:** 
- It seems that there is a noticeable positive relationship between `lead time` and `cancellations rate`. The higher the lead time, the greater the percentage of cancellations may be.

- This trend suggests that bookings made further in advance are more likely to be canceled compared to those made closer to the arrival date.

### 4. EDA 3D

#### 4.1. 3D data analysis for three numerical variables

First let's define the function `create_3d_scatter_plot` generates a 3D scatter plot using three specified features from a given dataset. It visualizes the relationships between these features in a 3D space, with additional surfaces representing the minimum values along each axis. 

```python
def create_3d_scatter_plot(feature1, feature2, feature3, data):
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Color by a fourth variable (if available)

    sc = ax.scatter(data[feature1], data[feature2], data[feature3], s=50, marker='o')

    ax.set_xlabel(feature1, fontsize=10, fontweight='bold')
    ax.set_ylabel(feature2, fontsize=10, fontweight='bold')
    ax.set_zlabel(feature3, fontsize=10, fontweight='bold')
    
    # ax.view_init(elev=0, azim=-0)

    plt.title(f'3D Scatter Plot of {feature1}, {feature2}, and {feature3}', fontsize=14, fontweight='bold')
    
    
    x = np.linspace(*ax.get_xlim(), 2)
    y = np.linspace(*ax.get_ylim(), 2)
    z = np.linspace(*ax.get_zlim(), 2)
    X, Z = np.meshgrid(x, z)
    Y = np.array([[y.min()]*len(x)]*len(z))
    ax.plot_surface(X, Y, Z, color='r', alpha=0.05)
    Y, Z = np.meshgrid(y, z)
    X = np.array([[x.min()]*len(y)]*len(z))
    ax.plot_surface(X, Y, Z, color='g', alpha=0.05)
    X, Y = np.meshgrid(x, y)
    Z = np.array([[z.min()]*len(x)]*len(y))
    ax.plot_surface(X, Y, Z, color='b', alpha=0.05)

    plt.show()
```

##### Booking changes, Weekend Stays, and Weeknight Stays

```python
create_3d_scatter_plot('booking_changes', 'stays_in_weekend_nights', 'stays_in_week_nights', df)
```

**Comments**:  
- It appears that there is a slight trend indicating that as the number of booking changes increases, guests tend to have fewer stays in weekend nights. Meanwhile, there seems to be a trend that as booking changes increase, guests tend to have higher stays in week nights.

- Individuals who tend to stay during week nights tend to have a higher proportion of stays during weekend nights.

##### Lead Time, Previous Cancellations, and Booking Changes

```python
create_3d_scatter_plot('lead_time', 'previous_cancellations', 'booking_changes', df)
```

**Comments**: 
- There is a slight inverse trend between previous_cancellations and lead_time. As lead_time increases, the proportion of previous_cancellations decreases significantly, almost to zero.

- It is evident that as lead_time increases, booking_changes tend to decrease significantly.
